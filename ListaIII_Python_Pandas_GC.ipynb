```python
# Importando biblioteca de trabalho
import pandas as pd
```

```python
# Lendo os dados e criando dataframe de trabalho
df_pre = pd.read_csv('https://raw.githubusercontent.com/guilhermecadori/NBADATA/main/nba.csv')

# Conferindo os dados importados
df_pre
```

### Limpando os dados

```python
# Conferindo NaN
na_Age = df_pre['Age'].isna().sum()
na_Height = df_pre['Height'].isna().sum()
na_Weight = df_pre['Weight'].isna().sum()

# Conferindo a quantidade de valores NaN encontrados
print('Quantidade NaN Age:', na_Age)
print('Quantidade NaN Height:', na_Height)
print('Quantidade NaN Weight:', na_Weight)
```

***Observação***

Será feita uma limpeza verificando apenas as colunas cujas médias devem ser calculadas, uma vez que se opta-se por utilizar a função dropna() no conjunto de dados como um todo, diversos registros seriam excluídos sem necessidade. Isso deve ao fato de a função dropna() sem a utilização do argumento *subset* exlcui qualquer registro que contenha NaN ou NA independente de qual coluna tal valor esteja.

```python
# df com Age limpo
# Remover NaN
df_Age = df_pre.dropna(subset = ['Age'])

df_Age
```

```python
# df com Height limpo
# Remover NaN
df_Height = df_pre.dropna(subset = ['Height'])

df_Height
```

```python
# df com Weight limpo
# Remover NaN
df_Weight = df_pre.dropna(subset = ['Weight'])

df_Weight
```

Apenas um registro em cada dataframe precisou ser removido. Sendo assim, um mesmo dataframe será utilizado para realizar os cálculos necessários, conforme abaixo.

```python
# Criando novo e único dataframe limpo

# Unindo dataframes df_Age e df_Height com base em seus valores em comum
merged_df1 = pd.merge(df_Age, df_Height,
                      on = ['Name', 
                            'Team', 
                            'Number', 
                            'Position', 
                            'Age', 
                            'Height',
                            'Weight',
                            'College',
                            'Salary'], 
                      how = 'inner')

# Unindo dataframes df_Weight e merged_df1 (criado no passo anterior) com base em seus valores em comum
df = pd.merge(df_Weight, merged_df1,
                      on = ['Name', 
                            'Team', 
                            'Number', 
                            'Position', 
                            'Age', 
                            'Height',
                            'Weight',
                            'College',
                            'Salary'], 
                      how = 'inner')

# Conferindo o novo dataframe limpo criado
df
```

## Utilizando a função crosstab() do pacote Pandas

### Calculando altura e idade médias dos jogadores para cada time

```python
# Calculando a média do peso dos jogadores de cada time
tabCrosstab = pd.crosstab(index = df['Team'], 
                          columns = ['Average Weight'], 
                          values = df['Weight'], 
                          aggfunc = 'mean')

# Calculando a média de idade dos jogadores de cada time
tabCrosstab2 = pd.crosstab(index = df['Team'], 
                           columns = ['Average Age'], 
                           values = df['Age'], 
                           aggfunc = 'mean')

# Unindo crosstabs anteriores
tabCrosstab3 = pd.merge(tabCrosstab2, 
                        tabCrosstab,
                        on = ['Team'], 
                        how = 'inner')

# Verificando resultados
tabCrosstab3
```

***Oberservação***

A função merge() foi utilizada devido à impossibilidade de criar médias para mais de uma coluna simultaneamente utilizando a função crosstab por si só.

### Calculando altura e idade médias dos jogadores para cada time E posição

```python
# Peso médio dos jogadores de cada time
tabCrosstab4 = pd.crosstab(index = [df['Team'], df['Position']], 
                           columns = ['Average Weight'], 
                           values = df['Weight'], 
                           aggfunc = 'mean')

# Idade média dos jogadores de cada time
tabCrosstab5= pd.crosstab(index = [df['Team'], df['Position']], 
                          columns = ['Average Age'],
                          values = df['Age'],
                          aggfunc = 'mean')

# Criando crosstab unindo crosstabs anteriores
tabCrosstab6 = pd.merge(tabCrosstab5,
                        tabCrosstab4,
                        on = ['Team', 'Position'], 
                        how = 'inner')

# Verificando resultados
tabCrosstab6
```

## Utilizando a função pivot_table() do pacote Pandas

### Calculando altura e idade médias dos jogadores para cada time

```python
# Agregando os dados e sumarizando os valores de peso e idade em valores médios
summary_pivottable = pd.pivot_table(df, 
                             index=['Team'], 
                             values=['Weight', 'Age'], 
                             aggfunc='mean').add_prefix('Average ')

# Verificando resultados
summary_pivottable
```

### Calculando altura e idade médias dos jogadores para cada time E posição.

```python
# Agregando os dados e sumarizando os valores de peso e idade em valores médios
summary_pivottable = pd.pivot_table(df, 
                             index=['Team', 'Position'], 
                             values=['Weight', 'Age'], 
                             aggfunc='mean').add_prefix('Average ')

# Verificando resultados
summary_pivottable
```
